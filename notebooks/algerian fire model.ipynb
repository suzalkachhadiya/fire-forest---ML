```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"region"]=0
df.loc[122:,"region"]=1
df1=df
```

```python
df1.head()
```

```python
df1[["region"]]=df1[["region"]].astype(int)
```

```python
df1.isnull().sum()
```

```python
df1=df1.dropna().reset_index(drop=True)
```

```python
df1.isnull().sum()
```

```python
df1.iloc[[122]]
```

```python
df1=df1.drop(122).reset_index(drop=True)
```

```python
df1.iloc[[122]]
```

```python
df1.columns=df1.columns.str.strip()
df1.columns
```

```python
df1[["day","month","year","Temperature","RH","Ws"]]=df1[["day","month","year","Temperature","RH","Ws"]].astype(int)
```

```python
df1.head()
```

```python
objects=[features for features in df1.columns if df1[features].dtypes == "O"]
```

```python
for i in objects:
    if i != "Classes":
        df1[i]=df1[i].astype(float)
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1.to_csv("algerian_forest_fire_cleanned_dataset.csv",index=False)
```
